```python
import pandas as pd
```

```python
data_csvs = ["1. final_dataset(0-200000).csv", 
            "2. final_dataset(200000-up).csv"]

li = []

for csv in data_csvs:
    data_df = pd.read_csv(csv, index_col=0, header=0)
    li.append(data_df)

full_data = pd.concat(li, axis=0, ignore_index=True)

full_data
```

```python
counts = full_data['Tract'].value_counts()

full_data = full_data[~full_data['Tract'].isin(counts[counts < 2].index)]

full_data
```

```python
small_df = full_data.dropna()

small_df['Tract'].value_counts()
```
