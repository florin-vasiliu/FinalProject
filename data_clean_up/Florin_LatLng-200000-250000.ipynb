```python
import pandas as pd
```

```python
latlng_df = pd.read_csv('revised_crimes_geo.csv')
latlng_df
```

```python
address_list = latlng_df['Full_Address'].to_list()
```

```python
from geopy.geocoders import Nominatim

crime_lats = []
crime_lngs = []
counter = 0

for address in address_list[200000:250000]:
    try:
        geolocator = Nominatim(user_agent="DreamTeam")
        location = geolocator.geocode(address)
#         print(location.raw)
        print(f'{counter}:{address} LatLng:{location.latitude, location.longitude}')
        print('----------------')
        crime_lats.append(location.latitude)
        crime_lngs.append(location.longitude)
    except:
        print(f'{counter}:Not Found! :(')
        print('----------------')
        crime_lats.append('NaN')
        crime_lngs.append('NaN')
    counter+=1

latlng_florin = pd.DataFrame(crime_lats, columns=['Latitude'])
latlng_florin['Longitude'] = crime_lngs

latlng_florin.to_csv('LatLng5.csv', index=False)
```
