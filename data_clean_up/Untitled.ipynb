```python
import pandas as pd
```

```python
lat_lngs = ["LatLng1.csv", 
            "LatLng2.csv",
            "LatLng3.csv",
            "LatLng4.csv",
            "LatLng5.csv",
            "LatLng6.csv",
            "LatLng7.csv",
            "LatLng8.csv",]

li = []

for csv in lat_lngs:
    df = pd.read_csv(csv, index_col=None, header=0)
    li.append(df)

full_lat_lngs = pd.concat(li, axis=0, ignore_index=True)

full_lat_lngs
```

```python
crime_df = pd.read_csv('revised_crimes_geo.csv')

crime_df
```

```python
crime_df['Latitude'] = full_lat_lngs['Latitude']

crime_df
```

```python
crime_df['Longitude'] = full_lat_lngs['Longitude']

crime_df
```

```python
census_csvs = ["census1.csv", 
            "census2.csv",
           "census3.csv",
           "census4.csv",
           "census5.csv",
           "census6.csv",
           "census7.csv",
           "census8a.csv",
           "census8b.csv",
           "census8c.csv",
           "census8d.csv",
           "census8e.csv",
           "census8f.csv",]

li = []

for csv in census_csvs:
    df = pd.read_csv(csv, index_col=None, header=0)
    li.append(df)

tract_df = pd.concat(li, axis=0, ignore_index=True)

tract_df
```

```python
crime_df['Tract'] = tract_df['Census_Tract']

crime_df
```

```python
crime_df.to_csv('Full_Crimes_CSV.csv')
```
