```python
# Import Dependencies
from sqlalchemy import create_engine
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String, Float, Date 
```

```python
# Create Schema
Base = declarative_base()

# Creates Classes which will serve as the anchor points for our Tables
class Cleaned_data(Base):
    __tablename__ = 'cleaned_data'
    id = Column(Integer, primary_key=True)
    Date = Column(Date)
    Hour = Column(Integer)
    Offense_Type = Column(String(255))
    Offense_Count = Column(String(255))
    Premise = Column(String(255))
    Block_Range = Column(String(255))
    Street_Name = Column(String(255))
    Street_Type = Column(String(255))
    City = Column(String(255))
    State = Column(String(255))
    Street_Address = Column(String(255))
    Full_Address = Column(String(255))
    Latitude = Column(Float)
    Longitude = Column(Float)
    Tract = Column(Integer)
    Year = Column(Integer)
    Month = Column(Integer)
    Day = Column(Integer)
    Week_Num = Column(Integer)
    Day_of_Week = Column(String(255))
    Temperature = Column(Float)
    Precipitation = Column(Float)
    Weather = Column(String(255))
    Weather_Description = Column(String(255))
    Total_Popupation = Column(Float)
    Total_Households = Column(Float)
    Median_Household_Income = Column(Float)
```

```python
# Create a session
from sqlalchemy.orm import Session
session = Session(bind=engine)
```

```python
# Build schema into DB
Base.metadata.create_all(engine)
```

```python
# Call the constructors and insert data
length = len(cleaned_data)
for index, row in cleaned_data.iterrows():
    instance = Cleaned_data(id=index, 
                            Date=row["Date"], 
                            Hour=row["Hour"], 
                            Offense_Type=row["Offense_Type"], 
                            Offense_Count=row["Offense_Count"], 
                            Premise=row["Premise"], 
                            Block_Range=row["Block_Range"], 
                            Street_Name=row["Street_Name"], 
                            Street_Type=row["Street_Type"], 
                            City=row["City"], 
                            State=row["State"], 
                            Street_Address=row["Street_Address"], 
                            Full_Address=row["Full_Address"], 
                            Latitude=row["Latitude"], 
                            Longitude=row["Longitude"], 
                            Tract=row["Tract"], 
                            Year=row["Year"], 
                            Month=row["Month"], 
                            Day=row["Day"], 
                            Week_Num=row["Week_Num"], 
                            Day_of_Week=row["Day_of_Week"], 
                            Temperature=row["Temperature"], 
                            Precipitation=row["Precipitation"], 
                            Weather=row["Weather"], 
                            Weather_Description=row["Weather_Description"], 
                            Total_Popupation=row["Total_Popupation"], 
                            Total_Households=row["Total_Households"], 
                            Median_Household_Income=row["Median_Household_Income"], 
                           )
    clear_output(wait=True)
    print(f"inserted index {index} from total {length}")
    try:
        session.add(instance)
        session.commit()
    except:
        session.rollback()
```
