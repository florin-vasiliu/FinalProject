```python
import pandas as pd
```

```python
houston_weather = pd.read_csv('Luis/houston_weather.csv')
crime_data = pd.read_csv('data_clean_up/Full_Crimes_CSV.csv')
```

```python
houston_weather = houston_weather.rename(columns={"Hour_Beginning":"Hour"})
houston_weather.head()
```

```python
crime_data.head()
```

```python
#sort crime df
crime_data = crime_data.sort_values(["Date", "Hour"])

#delete unecessary columns
del crime_data["Unnamed: 0"]
del crime_data["Unnamed: 0.1"]
del crime_data["Zip_Code"]

crime_data['Tract'] = crime_data['Tract'].apply(lambda f: format(f, '.0f'))
```

```python
crime_data['Year'] = pd.DatetimeIndex(crime_data['Date']).year
crime_data['Month'] = pd.DatetimeIndex(crime_data['Date']).month
crime_data['Day'] = pd.DatetimeIndex(crime_data['Date']).day
```

```python
crime_data.reset_index(drop=True)
```

```python
crime_data.dtypes
```

```python
crime_data['Hour'] = crime_data['Hour'].astype(int)
```

```python
crime_data.dtypes
```

```python
houston_weather.dtypes
```

```python
full_data = pd.merge(crime_data, houston_weather, on = ['Year', 'Month', 'Day', 'Hour'], how="left")
```

```python
del full_data["Date_Time"]
del full_data["Date_y"]
del full_data["Unnamed: 0"]
```

```python
full_data = full_data.rename(columns = {"Date_x":"Date"})
```
